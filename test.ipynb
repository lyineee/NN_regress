```python
import matplotlib.pyplot as plt 
from random import random

import numpy as np 
```

```python
def graph():
    x=np.linspace(-5,5,50)
    y=np.ones((50,))
    for i,data in enumerate(x):
        y[i]=x[i]**2 + 1*random() - 25
    plt.figure()
    plt.plot(x, y)
    plt.show()
```

```python
graph()
```
